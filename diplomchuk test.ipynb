```python
from ultralytics import YOLO
import time

# ---------------------
# 1. Обработка small моделью
# ---------------------
print("Обработка small моделью...")
model_small = YOLO('yolo_small_weights.pt')
start = time.time()
results_small = model_small('fight.mp4', save=True, conf=0.5)
end = time.time()
print(f"Small модель: {end - start:.2f} сек")
for r in results_small:
    if r.boxes is not None:
        print(f"  Найдено объектов: {len(r.boxes)}")
        for box in r.boxes:
            print(f"    Уверенность: {box.conf[0]:.2f}")

# ---------------------
# 2. Обработка nano моделью
# ---------------------
print("\nОбработка nano моделью...")
model_nano = YOLO('Yolo_nano_weights.pt')
start = time.time()
results_nano = model_nano('fight.mp4', save=True, conf=0.5)
end = time.time()
print(f"Nano модель: {end - start:.2f} сек")
for r in results_nano:
    if r.boxes is not None:
        print(f"  Найдено объектов: {len(r.boxes)}")
        for box in r.boxes:
            print(f"    Уверенность: {box.conf[0]:.2f}")
    else:
        print("  Не найдено ни одной драки")
```
