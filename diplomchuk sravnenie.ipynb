```python
from ultralytics import YOLO
import time

# Путь к видео
video_path = 'fight.mp4'

# Список моделей: (имя, путь к файлу)
models = [
    ('yolo_small_weights.pt', 'yolo_small_weights.pt'),
    ('Yolo_nano_weights.pt', 'Yolo_nano_weights.pt')
]

print("Сравнение моделей на видео:", video_path)
print("-" * 60)

results_data = []

for name, path in models:
    print(f"\nЗапуск модели {name}...")
    model = YOLO(path)
    
    start = time.time()
    results = model(video_path, conf=0.5, verbose=False)
    end = time.time()
    elapsed = end - start
    
    # Собираем информацию о найденных объектах
    total_boxes = 0
    confidences = []
    for r in results:
        if r.boxes is not None:
            total_boxes += len(r.boxes)
            confidences.extend(box.conf[0].item() for box in r.boxes)
    
    avg_conf = sum(confidences)/len(confidences) if confidences else 0
    
    results_data.append({
        'Модель': name,
        'Время (сек)': round(elapsed, 2),
        'Найдено драк': total_boxes,
        'Средняя уверенность': round(avg_conf, 2)
    })
    
    print(f"  Время: {elapsed:.2f} сек")
    print(f"  Найдено драк: {total_boxes}")
    print(f"  Средняя уверенность: {avg_conf:.2f}")

# Вывод итоговой таблицы
print("\n" + "="*60)
print("ИТОГОВОЕ СРАВНЕНИЕ")
print("-"*60)
print(f"{'Модель':<25} {'Время (сек)':<12} {'Найдено драк':<15} {'Ср. уверенность':<15}")
print("-"*60)
for row in results_data:
    print(f"{row['Модель']:<25} {row['Время (сек)']:<12} {row['Найдено драк']:<15} {row['Ср. уверенность']:<15}")
print("="*60)
```
